# Distributions  
Numpy: https://numpy.org/doc/stable/reference/random/generator.html#distributions  


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
seed = 1
rand = np.random.default_rng(seed)
```

### Binomial

```python
b = rand.binomial(n=1, p=0.2, size=100)

plt.hist(b);
```

### Beta

```python
bt = rand.beta(a=2, b=5, size=100)

plt.hist(bt);
```

### Normal

```python
n = rand.normal(loc=1, scale=0.2, size=100)

plt.hist(n);
```

### Uniform

```python
u = rand.uniform(low=0.0, high=1.0, size=100)

plt.hist(u);
```

### Exponentail

```python
e = rand.exponential(scale=1.0, size=100)

plt.hist(e);
```

### Lognormal

```python
l = rand.lognormal(mean=1, sigma=1, size=100)

plt.hist(l);
```

```python
plt.hist(np.log(l));
```

### Chisquare

```python
c = rand.chisquare(df=2, size=100)

plt.hist(c);
```

### Poisson

```python
p = rand.poisson(lam=1, size=100)

plt.hist(p);
```
